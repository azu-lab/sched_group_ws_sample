# 6. 2. evaluate with varing parameter
parameters are handed to src.simulate.evaluate_varing_parameter function as arguments

the return value is a dict type:

  evaluate_varing_parameter(samples, first_param=[...], second_param=[...], ...)

  -> results[first][second] = [...(result when first and second for all samples)]

```python
from src.simulate import evaluate_varing_parameter

results = evaluate_varing_parameter(dags, method=['proposed', 'existing'], core_num=[2, 4, 6])
print(results)
```

# 6. 3. visualize
ex) case of boxplot

```python
from matplotlib.pyplot import boxplot, show

for method in ['proposed', 'existing']:
    array = []
    for core in [2, 4, 6]:
        array.append(results[method][core])
    plot = boxplot(array, labels=[2, 4, 6])
    show()
```
